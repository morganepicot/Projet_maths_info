**Les fonctions**

```python
def grad(f):
    g = autograd.grad
    def grad_f(x, y):
        return np.array([g(f, 0)(x, y), g(f, 1)(x, y)])
    return grad_f

def J(f):
    j = autograd.jacobian
    def J_f(x, y):
        return np.array([j(f, 0)(x, y), j(f, 1)(x, y)]).T
    return J_f

def display_contour(f, x, y, levels):
    X, Y = np.meshgrid(x, y)
    Z = f(X, Y)
    fig, ax = plt.subplots()
    contour_set = plt.contour(
        X, Y, Z, colors="grey", linestyles="dashed", 
        levels=levels 
    )
    ax.clabel(contour_set)
    plt.grid(True)
    plt.xlabel("$x_1$") 
    plt.ylabel("$x_2$")
    plt.gca().set_aspect("equal")
```

### **Analyse des ensembles de niveau**

#### **Question 1**

On sait que $\lim\limits_{\|x \| \to +\infty} f(x) = +\infty$.  
Ainsi, pour $c$ fixé dans 
$
\mathbb{R}
$
, on peut trouver un réel A tel que : 
$$
\forall x=(x_1,x_2) \in \mathbb{R}^2 , \| x \| >A \implies f(x) > c
$$
  
La contraposée de cette propriéte implique donc que si $f(x) \leq c$, et en particulier $f(x) = c$, alors nécessairement, $\| x\| \leq A$.  
Ainsi, l'ensemble de niveau de $c$, noté $\mathscr{L}_c$ est borné par A.
De plus, on constate qu'il peut s'écrire $f^{-1} (\{c \})$.
C'est donc l'image réciproque d'un fermé par une fonction continue. On en déduit qu'il est fermé.  
Comme nous travaillons dans $\mathbb{R}$, les deux résultats précédents nous permettent de conclure d'après le théorème de Heine-Borel que $\mathscr{L}_c$ est un compact (partie fermée bornée de 
$\mathbb{R}$).

#### **Question 2**

On considère la matrice $A(x,y)$ définie par:
$$
A(x,y) = \begin{pmatrix}
\frac{(\nabla f)_1}{\| \nabla f\|} & x_1-x_{10} \\ \frac{(\nabla f)_2}{\| \nabla f\|} & x_2-x_{20}\end{pmatrix}
$$
On constate alors que 
$$
p(x_1,x_2) = \frac{\partial_2 f(x_0)}{\|\nabla f(x_0)\|} (x_1 - x_{10}) -
\frac{\partial_1 f(x_0)}{\|\nabla f(x_0)\|} (x_2 - x_{20}) = \mathrm{det}(A)
$$
  
$p(x_1,x_2)$ peut donc être défini comme étant l'aire algébrique du parallélogramme formé par les vecteurs $\frac{1}{\| \nabla f\|} \nabla f$ et $\begin{pmatrix} x_1 - x_{10} \\ x_2 - x_{20} \end{pmatrix}$.

#### **Question 3**

On cherche à utiliser le théorème des fonctions implicites pour montrer le résultat. On considère pour cela un ouvert $U$ où $\nabla f$ ne s'annule pas (qui existe bien par hypothèse) et la fonction $g :U \subset \mathbb{R}^2 \times \mathbb{R} \to \mathbb{R}^2$ telle que:
$$
\forall (x, t) \in U \times \mathbb{R} , g(x,t) = \begin{pmatrix} f(x_1,x_2) - c \\ t - p(x_1,x_2) \end{pmatrix}
$$
  
g est donc continûment différentiable par composée de fonctions continûment différentiables.
On différentie alors l'expression précédente par rapport à x :
$$
\partial_x g(x,t) = \begin{pmatrix} \partial_1 f(x) & \partial_2 f(x) \\ \frac{-\partial_2 f(x_0)}{\| \nabla f(x_0) \|} & \frac{\partial_1 f(x_0)}{\| \nabla f(x_0) \|} \end{pmatrix}
$$
On calcule le déterminant : $\: \mathrm{det}(\partial_x g(x,t)) = \frac{\langle \nabla f(x),\nabla f(x_0) \rangle}{\| \nabla f(x_0)\|}$
Cette valeur ne dépend que de $x$. De plus, elle est non nulle en $x_0$ (elle vaut $1$). Par continuité du déterminant, il existe un voisinage $V \subset U$ de $x_0$ tel que le déterminant ne s'annule pas, et alors $\partial_x g(x, t)$ est inversible pour tout $(x,t)$ de $V \times \mathbb{R}$. On se place sur cet ouvert pour la suite de la question. 
  
On a $g(x_0,0) = 0$ : $f(x_0) = c$ par hypothèse et $p(x_0) = 0$.
On peut alors appliquer le théorème des fonctions implicites à la fonction $g$.  
  
Soient $V_0 \subset V$ un voisinage de $x_0$, $W_0 \subset \mathbb{R}$ un voisinage de $0$, et une fonction $\gamma : W_0 \to \mathbb{R}^2$ continûment différentiable sur $W_0$ telle que :
$$
\forall x = (x_1, x_2) \in V_0, \forall t \in W_0, g(x, t) = 0 \iff x = \gamma (t)
$$
Seulement la condition $g(x,t) = 0$ implique que les deux composantes du vecteurs soient nulles. On peut la décomposer en $f(x) = c$ et $t = p(x)$.
  
En considérant $\epsilon >0$ tel que $]- \epsilon, \epsilon[ \subset W_0$ et en restreignant $\gamma$ à cet intervalle, on obtient finalement le résultat attendu.

#### **Question 4**

Le théorème des fonctions implicites nous donne une expression de la différentielle de $\gamma$ qui correspond ici à sa dérivée par rapport à $t$ (fonction d'une seule variable). On a :
$$
\forall t\in]-\epsilon,\epsilon[, \gamma'(t) = -(\partial_xg(\gamma(t),t))^{-1}. \partial_tg(\gamma(t),t)
$$
Or d'après la définition de g, on a $\partial_tg(x,t) = \begin{pmatrix} 0 \\ 1 \end{pmatrix}$. De plus, on connait l'expression de l'inverse d'une matrice $2 \times 2$ grâce à la méthode des cofacteurs : 
$$
(\partial_tg(x,t))^{-1} = \frac{\| \nabla f(x_0)\|}{\langle \nabla f(x),\nabla f(x_0) \rangle} \begin{pmatrix} \frac{\partial_1 f(x_0)}{\| \nabla f(x_0) \|} & -\partial_2 f(x) \\ -\frac{-\partial_2 f(x_0)}{\| \nabla f(x_0) \|} & \partial_1 f(x) \end{pmatrix}
$$
D'où $\forall t\in]-\epsilon,\epsilon[, \:\gamma '(t) = \frac{\| \nabla f(x_0)\|}{\langle \nabla f(\gamma(t)),\nabla f(x_0) \rangle} \begin{pmatrix} \partial_2f(\gamma(t)) \\ -\partial_1f(\gamma(t)) \end{pmatrix}$.
On a bien $\gamma'$ qui ne s'annule pas sur $]-\epsilon,\epsilon[$ car $\nabla f$ est non nul sur $V_0$.  
  
De plus, $\langle \nabla f(\gamma(t)), \gamma'(t) \rangle = \partial_1 f(\gamma(t)) \times \partial_2f(\gamma(t)) - \partial_2 f(\gamma(t)) \times \partial_1f(\gamma(t)) = 0$.  
  
Le vecteur $\gamma'(t)$ est bien non nul et orthogonal à $\nabla f(\gamma(t))$ sur $]-\epsilon,\epsilon[$.

### Méthode de Newton

#### **Question 5**

On a besoin d'avoir une courbe de niveau cohérente et donc que le couple $(x_1, x_2)$ ne s'éloigne "pas trop" de la valeur exacte pour chaque solution de $f(x_1,x_2)=c$. De plus la méthode de Newton converge rapidement donc choisir un $\epsilon$ trop petit n'apporte pas de précision supplémentaire. 
On considère donc $\epsilon = 1.10^{-5}$.

```python
N = 100
eps = 1e-5
```

#### **Tâche 1**

```python
def Newton(F, x0, y0, eps=eps, N=N):
    for i in range(N):
        diff_inv = np.linalg.inv(J(F)(x0, y0))
        x, y = np.array([x0, y0]) - np.dot(diff_inv, F(x0, y0)) #on applique
                                                                #le schéma de récurrence de Newton
        if np.sqrt((x - x0)**2 + (y - y0)**2) <= eps:
            return x, y
        x0, y0 = x, y
    raise ValueError(f"no convergence in {N} steps.")
```

### Génération des points

#### **Question 6 + Tâche 3**

La question nous invite à utiliser la fonction Newton précédemment définie. En effet, pour trouver $(x,y)$ satisfaisant les deux conditions énoncées, nous pouvons définir une fonction $F :\mathbb{R}^2 \to \mathbb{R}^2$ telle que :

$$
\forall (x,y) \in \mathbb{R}^2 , F(x,y) = \begin{pmatrix} f(x,y) - f(x_0,y_0) \\ (x-x_0)^2 + (y-y_0)^2 - \delta \end{pmatrix}
$$  
Ainsi, le point obtenu après convergence du schéma sera sur la même ligne de niveau que $(x_0,y_0)$ et sera sur le cercle de centre $(x_0,y_0)$ et de rayon $\delta$.  
  
    

Mais avant de chercher le point de convergence, on veut que celui-ci converge vers la solution à droite de $\nabla f(x_0,y_0)$. Ainsi, on veut que le point initial en argument de la fonction Newton soit suffisamment proche de la solution pour converger vers celle-ci (et non pas vers l'autre solution située à gauche du gradient).  
Pour cela, on introduit une troisième dimension qui va nous permettre d'orienter l'espace à l'aide du produit vectoriel. On définit les vecteurs :
$$
gradf_{3D} = \begin{pmatrix} {\nabla f(x_0,y_0)}_1 \\{\nabla f(x_0,y_0)_2} \\ 0 \end{pmatrix}
$$  

$$
Z_{3D} = \begin{pmatrix} 0 \\ 0 \\ 1 \end{pmatrix}
$$  

On effectue leur produit vectoriel et on le normalise : soit $W$ tel que $W = \frac{gradf_{3D} \land Z_{3D}} {\| gradf_{3D} \land Z_{3D} \|}$.  
Ce vecteur unitaire est orthogonal à $\nabla f(x_0,y_0)$ et dirigé à sa droite.
En prenant pour point initial le couple $(x_1,y_1)$ tel que :
$$
(x_1,y_1) = (x_0,y_0) + \delta(W_1,W_2)
$$
On est assuré de converger vers le point voulu.  
  
    
    
En appliquant le raisonnement précedent, on obtient la fonction suivante :

```python
def level_curve(f, x0, y0, delta = 0.1, N = 50, eps = eps):
    
    def F(x, y):
        return np.array([f(x, y) - f(x0, y0), 
                         (x - x0)**2 + (y - y0)**2 - delta**2])
    
    couples = np.zeros((2,N))
    
    for i in range(N):
        couples[0, i], couples[1, i] = x0, y0
        
        grad_f_3D = np.array([grad(f)(x0, y0)[0], grad(f)(x0, y0)[1], 0])
        Z_3D = np.array([0, 0, 1])
        W = np.cross(grad_f_3D, Z_3D)
        U = (1 / np.linalg.norm(W)) * W
        x1, y1 = x0 + delta * U[0], y0 + delta * U[1]
        x, y = Newton(F, x1, y1, eps)
        x0, y0 = x, y
    
    return couples
```

### Bouclage et auto-intersection

#### **Question 7 + Tâche 4**

Lors de l'implémentation de la nouvelle fonction level_curve (renommée level_curve_bis), il est nécessaire de savoir si on teste l'intersection du nouveau segment avec seulement le premier ce qui signifierait que les points ont fait le "tour" de la ligne de niveau, ou s'il faut tester avec tous les segments précédents. Cela peut être utile dans le cas de lignes de niveau comme celle de Rosenbrock prise comme exemple, mais entraine une augmentation importante de la complexité de l'algorithme.  
  
    
    
Cette implémentation ne sera de plus utile que si ce type d'auto-intersection est fréquente. Pour cela, il faut que la fonction étudiée ait son gradient qui s'annule au moins trois fois en des points distincts de $\mathbb{R}^2$, créant un effet de double bosse ou double creux séparés par un point col. De plus, pour ce type de fonction, seule la ligne de niveau associée au point d'annulation du gradient n'étant pas un extremum local sera auto-intersécante en ce point.  
Le cas semble ainsi suffisamment rare pour ne considérer que l'intersection entre le nouveau segment et le premier formé,  
ce que nous ferons dans la tâche 4.  
  
    
    
On ajoute tout de même une condition d'arrêt dans le cas où cela se produit pour que notre programme ne tourne pas à l'infini.  
  
  
    
      
    
Nous avons traduit la condition d'intersection des segments comme une condition de distance entre le nouveau point formé et le point de référence. Si celle-ci est inférieure à $\delta - \epsilon$, alors cela signifie que le prochain point sera situé après le premier, et donc que les segments vont s'intersecter.  
(On constate qu'avec cette méthode, on doit choisir judicieusement le point initial pour ne pas avoir de cas où la courbe de niveau passe suffisamment proche d'elle-même à cet endroit pour, sans s'auto-intersecter, être à une distance inférieure à $\delta - \epsilon$ d'elle-même.)

```python
def level_curve_bis(f, x0, y0, delta = 0.1, N = 50, eps = eps):
        
    def F(x, y):
        return np.array([f(x, y) - f(x0, y0), 
                         (x - x0)**2 + (y - y0)**2 - delta**2])
    def distance(x1, y1, x2, y2):
        return np.sqrt((x1 - x2)**2 + (y1 - y2)**2)
    
    absc, ordo = [], []
    x_ref, y_ref = x0, y0
    dist = delta + 1
    i = 0
    while dist >= delta - eps and i < 500: 
        
        #dans le pire des cas, le point trouvé 
        #est à une distance epsilon de celui sur la ligne de niveau
        #s'il y a un problème (fonction de Rosenbrock), on définit une condition d'arrêt
        #pour que l'algorithme ne tourne pas à l'infini
        
        #on fait le même raisonnement que pour la question précédente
        
        i += 1
        absc.append(x0)
        ordo.append(y0)
        grad_f_3D = np.array([grad(f)(x0, y0)[0], grad(f)(x0, y0)[1], 0])
        Z_3D = np.array([0, 0, 1])
        W = np.cross(grad_f_3D, Z_3D)
        U = (1 / np.linalg.norm(W)) * W
        
        x1, y1 = x0 + delta * U[0], y0 + delta * U[1]
        
        x, y = Newton(F, x1, y1, eps)
        x0, y0 = x, y
        
        dist = distance(x0, y0, x_ref, y_ref)
    return np.array([absc, ordo])
```

### Interpolation

#### **Question 8**

On considère les points $P_1 = (x_1,y_1)$ et $P_2 = (x_2, y_2)$ ainsi que les vecteurs $u_1 = \begin{pmatrix} x_3 \\ y_3 \end{pmatrix}$ et $u_2 = \begin{pmatrix} x_4 \\ y_4 \end{pmatrix}$.  
  
  
Comme $\gamma : t \to (x(t), y(t)) = (a + bt + ct^2, d +et + ft^2)$ est dérivable sur $\mathbb{R}$, on commence par déterminer sa dérivée :  
  
  
$$
\forall t \in \mathbb{R}, \gamma'(t) = (x'(t), y'(t)) = (b+2ct, e+2ft)
$$

De plus, la condition "$\gamma'(0)$ soit dans la même direction et le même sens que $u_1$ et $\gamma'(1)$ soit dans la même direction et le même sens que $u_2$" implique l'existence de deux réels strictement positifs (on ne peut pas diriger la courbe selon un vecteur nul) $\lambda_1$ et $\lambda_2$ tels que $\gamma'(0) = \lambda_1 u_1$ et $\gamma'(1) = \lambda_2 u_2$.
On obtient le système :  
  
    
    
$$
\begin{cases}
\gamma(0)= P_1 \\ \gamma(1)= P_2 \\\gamma'(0)=\lambda_1 u_1 \\\gamma'(1) = \lambda_2 u_2
\end{cases}
\; \; \; \; \Leftrightarrow \; \; \; \;
\begin{cases}
a = x_1 \; \; (1)\\
d = y_1 \; \;  (2)\\
a + b + c = x_2 \; \;  (3)\\
d + e + f = y_2 \; \;  (4)\\
b = \lambda_1 x_3 \; \;  (5)\\
e = \lambda_1 y_3 \; \;  (6)\\
b + 2c = \lambda_2x_4 \; \;  (7)\\
e + 2f = \lambda_2y_4 \; \;  (8)
\end{cases}
$$  
  
    
      
      
On trouve immédiatement $a=x_1$ et $d = y_1$. Puis on cherche à éliminer $\lambda_1$ et $\lambda_2$ afin de déterminer $b, c, e, f$.  
En réinjectant $(5)$ dans $(3)$ on trouve : $c = x_2 - x_1 -\lambda_1 x_3$  
De même, en réinjectant $(6)$ dans $(4)$ on trouve : $f = y_2 - y_1 -\lambda_1y_3$.  
Puis à l'aide de $(7)$ et $(8)$ on a :
$$
\begin{cases}
c = \frac{1}{2} (\lambda_2x_4 - \lambda_1x_3)\\
f = \frac{1}{2} (\lambda_2y_4 - \lambda_1y_3)
\end{cases}
$$  
  
  
Alors, en réinjectant les expressions trouvées dans les équations $(3)$ et $(4)$, on a éliminé toutes les inconnues autres que $\lambda_1$ et $\lambda_2$.
On a le système :
  
    
      
        
        
$$
\begin{cases}
x_2 - x_1 = \frac{1}{2} (\lambda_1 x_3 + \lambda_2 x_4)\\
y_2 - y_1 = \frac{1}{2} (\lambda_1 y_3 + \lambda_2 y_4)
\end{cases}
\Leftrightarrow
P_2 - P_1 = \frac{1}{2}(\lambda_1u_1 + \lambda_2u_2)
$$  
  
    
      
        
        
Ce système est de deux équations, deux inconnues. Il n'admettra une solution que si le déterminant de la matrice A associée est non nul. Or ici $A =\frac{1}{2} \times \begin{pmatrix} x_3 & x_4 \\ y_3 & y_4 \end{pmatrix}$, on a $det(A) = \frac{1}{4} (x_3y_4 - y_3x_4)$.  
  
    
    
**Ainsi, la première condition sera que $u_1$ et $u_2$ ne doivent pas être colinéaires**.

Puis lorsque c'est le cas, après de nouveaux calculs, on trouve finalement :  
  
  
  
 
$$
\begin{cases}
\lambda_1= \frac{2}{det(A)}((x_2-x_1)y_4 - (y_2-y_1)x_4)\\
\lambda_2 = \frac{2}{det(A)} ((y_2-y_3)x_3 - (x_2-x_1)y_3)
\end{cases}
$$  
  
    
      
      
  
Or on veut que $\lambda_1$ et $\lambda_2$ soient strictement positifs. On a deux cas de figure.
* **Si $det(A) > 0$**, alors les deux autres grandeurs doivent l'être aussi. Or pour $\lambda_1$, on reconnait $det(P_2 - P_1, u_2)$, et pour $\lambda_2$,  $det(P_2 - P_1, u_1)$. Ces deux valeurs pouvant être associées à l'aire algébrique du parallélogramme formé par $P_2 - P_1$ et $u_1$ ou respectivement $u_2$. **La deuxième condition sera que ces aires doivent être positives.** 
  
* **Si $det(A) < 0$, alors la deuxième condition sera que ces aires doivent être négatives**.
  
    

Si cette deuxième condition est vérifiée, il suffit alors de remplacer ces deux expressions dans celles de $b, c, e, f$ établies précédemment pour avoir $\gamma$.

#### Tâche 6

```python
# On définit une fonction auxiliaire qui teste si les conditions définies précedemment sont vérifiées.

def condition(P1, P2, u1, u2):
    d = u1[0] * u2[1] - u2[0] * u1[1]
    
    if d == 0 or u1[0] == 0:
        return False
    
    else:
        l2 = (2 * ((P1[0] - P2[0]) * u1[1] + (P2[1] - P1[1]) * u1[0])) / d
        l1 = (2 * (P2[0] - P1[0]) - l2 * u2[0]) / u1[0]
        
        if l2 <= 0 or l1 <= 0:
            return False
        
        else:
            return [l1, l2]



def gamma(t, P1, P2, u1, u2):
    P1 = P1.reshape(2, 1) # On reshape les vecteurs pour éviter les erreurs liées au broadcasting
    P2 = P2.reshape(2, 1)
    u1 = u1.reshape(2, 1)
    u2 = u2.reshape(2, 1)
    l = condition(P1, P2, u1, u2)
    
    if not l:
        return t * (P2 - P1) + P1

    else:
        a, d = P1
        b, e = l[0] * u1
        c, f = P2 - P1 - l[0] * u1
        A, B, C = np.array([a, d]), np.array([b, e]), np.array([c, f])
        return A + t * B + (t**2) * C

#On teste gamma sur un exemple arbitraire
ax = plt.axes()
G = gamma(np.linspace(0, 1, 100), np.array([3, 4]), np.array([11, 9]), np.array([1, -1]), np.array([1, 1]))
plt.plot(G[0], G[1])

#On trace les vecteurs pour vérifier qu'ils sont tangents à gamma
ax.arrow(3, 4, 1, -1, head_width = 0.1, head_length = 0.2, fc = 'red', ec = 'red')
ax.arrow(11, 9, 1, 1, head_width = 0.1, head_length = 0.2, fc = 'red', ec = 'red')

plt.xticks(np.arange(2, 14))
plt.yticks(np.arange(2, 12))

plt.show()
```

#### **Tâche 7**

```python
def level_curve3(f, x0, y0, delta, N, eps, oversampling):
    if oversampling == 1:
        return level_curve(f, x0, y0, delta, N, eps)
    
    elif oversampling > 1:
        K = oversampling - 1
        points = np.zeros((2, N + (N - 1) * K))

        M = level_curve(f, x0, y0, delta, 1, eps)

        points[0][0] = M[0]
        points[1][0] = M[1]
        
        for i in range(1, N):
            P1 = points[:, (i - 1) * (K + 1)]
            x, y = P1
            J = level_curve(f, x, y, delta, 2, eps) #On calcule P2
            P2 = np.array([J[0][1], J[1][1]])

            #On calcule les vecteurs u1 et u2 en prenant des orthogonaux au gradient
            Z_3D = np.array([0, 0, 1])
            g1 = np.array([grad(f)(x, y)[0], grad(f)(x, y)[1], 0])
            W1 = np.cross(g1, Z_3D)
            u1 = W1[:-1]

            g2 = np.array([grad(f)(P2[0], P2[1])[0], grad(f)(P2[0], P2[1])[1], 0])
            W2 = np.cross(g2, Z_3D)
            u2 = W2[:-1]

            
            t = np.linspace(0, 1, K) 

            T = gamma(t, P1, P2, u1, u2)
            
            #On ajoute les points calculés au tableau 
            points[:, (i - 1) * (K + 1) + 1: (i - 1) * (K + 1) + 1 + K] = np.copy(T)
            points[:, i * (K + 1)] = np.copy(P2)
                          
        return points
```
